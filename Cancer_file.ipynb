```python
# Read the text file and convert it to a CSV file
input_file = '/Users/sakshihegde/Documents/MS DS UoR /Sem 1/DSCC 440 (Data Mining)/Project 1/responses_Kamila_1.txt'
output_file = '/Users/sakshihegde/Documents/MS DS UoR /Sem 1/DSCC 440 (Data Mining)/Project 1/data_again3.csv'
```

```python
output_file
```

```python
# Open the text file for reading
with open(input_file, 'r') as infile:
    # Read all lines from the text file
    lines = infile.readlines()

# Open the CSV file for writing
with open(output_file, 'w', newline='') as outfile:
    # Create a CSV writer object with tab delimiter
    csv_writer = csv.writer(outfile, delimiter=',')  # You can use ',' or any separator
    
    # Loop through each line in the text file
    for line in lines:
        # Strip any leading/trailing whitespaces and split the line by tabs
        row = line.strip().split('\t')  # Split by tab
        # Write the row to the CSV file
        csv_writer.writerow(row)

print(f"Data has been converted to {output_file}")
```
